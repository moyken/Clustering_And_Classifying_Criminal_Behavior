# Define function to get suspect and victim name from case name

```python
def get_suspect(case_name):
    suspect = re.sub('\(s\)','',re.sub(' \(v\)','',re.findall('^\(s\).*\(v\)', case_name)[0]))
    return suspect
```

```python
def get_victim(case_name):
    victim = re.sub('\(v\) ','',re.findall('\(v\).*', case_name)[0])
    return victim
```

```python
DA_Full['suspect_name'] = DA_Full.CaseName.apply(lambda x: get_suspect(x))
DA_Full['victim_name'] = DA_Full.CaseName.apply(lambda x: get_victim(x))
```

# PCA Data Transformation

```python
suspect_actions.tail()
```

```python
pca = PCA()
pca.set_params(n_components = 36)
pca.fit(suspect_actions)
pca.components_
```

```python
f, ax = plt.subplots(figsize=(10, 10))
print(pca.explained_variance_ratio_.cumsum())
ax.plot(range(1,37),pca.explained_variance_ratio_.cumsum())
```

### Setting 36 Pricipal Components

Our goal isnt to reduce dimension, but to replace the binary values with weights

```python
pca.set_params(n_components = 36)
pca.fit(suspect_actions)
len(pca.components_)
```

```python
# 'componenets' are technically the loading vectors?
components = pd.DataFrame(pca.components_)
print('checking if squared sum is 1:',np.sum(components.iloc[0,:]**2))
components.tail()
```

```python
suspect_actions_pca = pd.DataFrame(pca.transform(suspect_actions))
suspect_actions_pca.tail()
```

# Skree plot to find best cluster size

```python
kmeans = KMeans()

inertia = []

for i in range(2,11):
    kmeans.set_params(n_clusters = i)
    kmeans.fit(suspect_actions_pca)
    inertia.append(kmeans.inertia_)
    
# Skree Plot to find best cluster size 
f, ax = plt.subplots(figsize=(10, 10))
ax.plot(range(2,11),inertia, marker = 'o')
```

# Refitting 5 Clusters. Best Within Cluster Variance is ~1260

5 clusters show the most unique and distinguishable groups. With 6, there are 2 clusters that are very similar

```python
kmeans.set_params(n_clusters = 5, n_init = 200)
kmeans.fit(suspect_actions_pca)
print('Table remains shape of 1185,1:',kmeans.labels_.shape)
print('Within Cluster Variance:', kmeans.inertia_)
```

# Concatenating all the DFs

```python
# cluster labels start at 0. Add 1 for readability
kmeans_clusters = pd.DataFrame(kmeans.labels_) + 1

# combine cluster labels with the original data set (with case info in front for convenience)
k_clusters = pd.concat([kmeans_clusters,case_info, DA_Full], axis = 1)
k_clusters.rename(columns = {0:'Cluster'}, inplace = True)
k_clusters.drop(case_info, axis = 1).drop(['DIRVersion','CaseName'], axis = 1).head()
```

### Ensuring the same data type

```python
k_clusters.iloc[:,8:47] = k_clusters.iloc[:,8:47].astype('int64')
```

# KMeans Cluster 1 - "Other" : Relatively Less Harmful

```python
## Subsetting Cluster 1
k_cluster_1 = k_clusters.loc[k_clusters.Cluster == 1,:]
k_cluster_1.to_csv('Cluster 1.csv')

# summary
cases = k_cluster_1.shape[0]
print(cases, 'incidents')
k_cluster_1 = pd.DataFrame(k_cluster_1.iloc[:,8:47].sum()).sort_values(by = 0, ascending = False)
k_cluster_1.rename(columns = {0:'Count'}, inplace = True)
k_cluster_1['Perc'] = (k_cluster_1.Count / cases)
print(k_cluster_1.Count.sum(), 'counts of crimes')
k_cluster_1.head(10)
```
